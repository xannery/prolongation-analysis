```python
import pandas as pd 
import numpy as np 
```

```python
df_fin = pd.read_csv('financial_data.csv')
df_pro = pd.read_csv('prolongations.csv')
```

```python
print("financial_data shape:", df_fin.shape)
print("prolongations shape:", df_pro.shape)

print("\n financial_data.info()")
df_fin.info()

print("\n prolongations.info()")
df_pro.info()

print("\n financial_data.head(10)")
display(df_fin.head(10))

print("\n prolongations.head(10)")
display(df_pro.head(10))
```

```python
month_mapping = {
    'ноябрь 2022': 'Ноябрь 2022',
    'декабрь 2022': 'Декабрь 2022',
    'январь 2023': 'Январь 2023',
    'февраль 2023': 'Февраль 2023',
    'март 2023': 'Март 2023',
    'апрель 2023': 'Апрель 2023',
    'май 2023': 'Май 2023',
    'июнь 2023': 'Июнь 2023',
    'июль 2023': 'Июль 2023',
    'август 2023': 'Август 2023',
    'сентябрь 2023': 'Сентябрь 2023',
    'октябрь 2023': 'Октябрь 2023',
    'ноябрь 2023': 'Ноябрь 2023',
    'декабрь 2023': 'Декабрь 2023',
}

df_pro['last_month_fin'] = df_pro['month'].map(month_mapping)
```

```python
df_merged = pd.merge(
    df_fin, 
    df_pro[['id', 'AM', 'last_month_fin']], 
    on='id', 
    how='left' 
)

print("Размер после merge:", df_merged.shape)
print("\nПервые 10 строк (только нужные колонки):")
display(df_merged[['id', 'AM', 'last_month_fin', 'Account']].head(10))

print("\nСколько проектов без last_month_fin?")
print(df_merged['last_month_fin'].isna().sum())
```

```python
missing = df_merged[df_merged['last_month_fin'].isna()][['id', 'Account', 'AM', 'last_month_fin']]
print("Проекты без last_month_fin:")
display(missing)
```

```python
df_clean = df_merged[df_merged['last_month_fin'].notna().copy()]

print("Размер после фильтрации:", df_clean.shape)
print("Сколько осталось NaN в last_month_fin?", df_clean['last_month_fin'].isna().sum())
```

```python
df_clean = df_clean.copy()  
df_clean['is_excluded'] = False

df_clean['is_excluded'] = df_clean.apply(
    lambda row: pd.notna(row['last_month_fin']) and 
                str(row[row['last_month_fin']]).strip() in ['стоп', 'end'],
    axis=1
)

print("Распределение is_excluded:")
print(df_clean['is_excluded'].value_counts())

print("\nПримеры исключённых проектов (первые 8):")
display(df_clean[df_clean['is_excluded']][['id', 'AM', 'last_month_fin', 'Ноябрь 2022', 'Декабрь 2022']].head(8))
```

```python
def clean_value(x):
    if pd.isna(x):
        return np.nan
    

    x_str = str(x).strip().replace('\xa0', '')
    
    if 'стоп' in x_str.lower() or 'end' in x_str.lower():
        return np.nan
    if 'в ноль' in x_str.lower():
        return 0.0
    
    try:
        num_str = x_str.replace(' ', '').replace(',', '.')
        return float(num_str)
    except ValueError:
        return np.nan
    
month_cols = [col for col in df_clean.columns if '202' in col]
df_clean[month_cols] = df_clean[month_cols].map(clean_value)

print("Проверка после очистки:")
display(df_clean[month_cols].head(10))

print("\nПример проекта с 'в ноль':")
display(df_clean[df_clean['id'] == 579][['id', 'AM', 'last_month_fin', 'Ноябрь 2022']].head())
```

```python
# 1. Агрегация по проектам
df_projects = df_clean.groupby(
    ['id', 'AM', 'last_month_fin', 'is_excluded'], 
    as_index=False
)[month_cols].sum()

print("Размер после агрегации (одна строка = один проект):", df_projects.shape)
print("\nПервые 10 проектов:")
display(df_projects.head(10))

# 2. Проверка, что is_excluded теперь на уровне проекта
print("\nСколько проектов исключено:")
print(df_projects['is_excluded'].value_counts())
```

```python
months_2023 = [
    'Январь 2023', 'Февраль 2023', 'Март 2023', 'Апрель 2023',
    'Май 2023', 'Июнь 2023', 'Июль 2023', 'Август 2023',
    'Сентябрь 2023', 'Октябрь 2023', 'Ноябрь 2023', 'Декабрь 2023'
]
results = []

for i, m in enumerate(months_2023):
    # Предыдущие месяцы
    prev_month = months_2023[i-1] if i > 0 else None
    prev2_month = months_2023[i-2] if i > 1 else None
    
    # ПЕРВЫЙ КОЭФФИЦИЕНТ
    if prev_month is None:
        first_base = 0
        first_prolonged = 0
    else:
        mask1 = (df_projects['last_month_fin'] == prev_month) & \
                (df_projects['is_excluded'] == False)
        
        first_base = df_projects.loc[mask1, prev_month].sum()
        first_prolonged = df_projects.loc[mask1, m].sum()
    
    first_coeff = first_prolonged / first_base if first_base > 0 else 0
    
    # ВТОРОЙ КОЭФФИЦИЕНТ
    if prev2_month is None:
        second_base = 0
        second_prolonged = 0
    else:
        mask2 = (df_projects['last_month_fin'] == prev2_month) & \
                (df_projects['is_excluded'] == False) & \
                (df_projects[prev_month] == 0)        
        second_base = df_projects.loc[mask2, prev2_month].sum()
        second_prolonged = df_projects.loc[mask2, m].sum()
    
    second_coeff = second_prolonged / second_base if second_base > 0 else 0
    
    results.append({
        'Месяц': m,
        'к пролонгации (1й мес)': round(first_base, 2),
        'пролонгировано (1й мес)': round(first_prolonged, 2),
        'Коэффициент (1й мес)': round(first_coeff, 4),
        'к пролонгации (2й мес)': round(second_base, 2),
        'пролонгировано (2й мес)': round(second_prolonged, 2),
        'Коэффициент (2й мес)': round(second_coeff, 4),
    })

df_report = pd.DataFrame(results)
display(df_report)
```

```python
# Весь отдел
total_row = {
    'Месяц': 'Весь отдел',
    'к пролонгации (1й мес)': round(df_report['к пролонгации (1й мес)'].sum(), 2),
    'пролонгировано (1й мес)': round(df_report['пролонгировано (1й мес)'].sum(), 2),
    'Коэффициент (1й мес)': round(
        df_report['пролонгировано (1й мес)'].sum() / df_report['к пролонгации (1й мес)'].sum(), 4
    ) if df_report['к пролонгации (1й мес)'].sum() > 0 else 0,
    'к пролонгации (2й мес)': round(df_report['к пролонгации (2й мес)'].sum(), 2),
    'пролонгировано (2й мес)': round(df_report['пролонгировано (2й мес)'].sum(), 2),
    'Коэффициент (2й мес)': round(
        df_report['пролонгировано (2й мес)'].sum() / df_report['к пролонгации (2й мес)'].sum(), 4
    ) if df_report['к пролонгации (2й мес)'].sum() > 0 else 0,
}

df_final = pd.concat([df_report, pd.DataFrame([total_row])], ignore_index=True)

display(df_final)
```

```python
# 1. Берём только активные проекты
df_active = df_projects[~df_projects['is_excluded']].copy()

# 2. Основная таблица — количество проектов
df_managers_year = (
    df_active
    .groupby('AM', as_index=False)
    .agg(projects=('id', 'count'))
)

# 3. Считаем сумму отгрузки за весь 2023 год
shipment_sums = (
    df_active
    .groupby('AM')[month_cols]
    .sum()
    .sum(axis=1)
    .reset_index(name='Сумма отгрузки 2023')
)

# 4. Соединяем
df_managers_year = df_managers_year.merge(shipment_sums, on='AM')

# 5. Приводим в красивый вид
df_managers_year = df_managers_year.rename(columns={'projects': 'Кол-во проектов'})
df_managers_year = df_managers_year.sort_values('Кол-во проектов', ascending=False).round(2)

display(df_managers_year)

print(f"\nВсего менеджеров: {len(df_managers_year)}")
print(f"Всего проектов по всем менеджерам: {df_managers_year['Кол-во проектов'].sum()}")
```

### ТАБЛИЦА "МЕНЕДЖЕРЫ ПО МЕСЯЦАМ"

```python
manager_monthly = []

for am in sorted(df_projects['AM'].unique()):
    df_am = df_projects[(df_projects['AM'] == am) & (~df_projects['is_excluded'])].copy()
    
    if len(df_am) == 0:
        continue
    
    for i, m in enumerate(months_2023):
        prev_month = months_2023[i-1] if i > 0 else None
        prev2_month = months_2023[i-2] if i > 1 else None
        
        # 1-й коэффициент
        if prev_month is None:
            first_base = 0
            first_prolonged = 0
        else:
            mask1 = (df_am['last_month_fin'] == prev_month)
            first_base = df_am.loc[mask1, prev_month].sum()
            first_prolonged = df_am.loc[mask1, m].sum()
        
        first_coeff = first_prolonged / first_base if first_base > 0 else 0
        
        # 2-й коэффициент
        if prev2_month is None:
            second_base = 0
            second_prolonged = 0
        else:
            mask2 = (df_am['last_month_fin'] == prev2_month) & (df_am[prev_month] == 0)
            second_base = df_am.loc[mask2, prev2_month].sum()
            second_prolonged = df_am.loc[mask2, m].sum()
        
        second_coeff = second_prolonged / second_base if second_base > 0 else 0
        
        manager_monthly.append({
            'AM': am,
            'Месяц': m,
            'к пролонгации (1й мес)': round(first_base, 2),
            'пролонгировано (1й мес)': round(first_prolonged, 2),
            'Коэффициент (1й мес)': round(first_coeff, 4),
            'к пролонгации (2й мес)': round(second_base, 2),
            'пролонгировано (2й мес)': round(second_prolonged, 2),
            'Коэффициент (2й мес)': round(second_coeff, 4),
        })
        
df_managers_monthly = pd.DataFrame(manager_monthly)

df_managers_monthly = df_managers_monthly.sort_values(['AM', 'Месяц'])

display(df_managers_monthly.head(20)) 
print(f"\nРазмер таблицы: {df_managers_monthly.shape}")
```
